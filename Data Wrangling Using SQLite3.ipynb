Now,we've a table with all the buildings in Gorkha, but the final piece of data needed for our model, the damage that each building sustained in the earthquake, is in the damage_grade table.Let's look at it.

### damage_grade

```python
%sql PRAGMA table_info(damage_grade);
```

I find the column we intersted in  damage_grade table which is   damage_grade column .

We can get the column we are interested in by joining the damage_grade table with the building_structure column but let's see what is the type of join we should use.

```python
%%sql
SELECT *
FROM damage_grade
WHERE district_id=36
LIMIT 5
```

```python
%%sql
SELECT count(*)
FROM damage_grade
WHERE district_id=36
LIMIT 5
```

 As I have two tables with the same number of rows and I want to combine the data based on matching rows, I can use an inner join.

### Inner join

```python
%%sql
SELECT distinct(s.building_id) AS b_id,
s.*,d.damage_grade
FROM building_structure AS s
INNER JOIN damage_grade AS d ON s.building_id = d.building_id
WHERE s.district_id = 36
LIMIT 5
```

## Import

```python
#connect to the database.
conn = sqlite3.connect("E:/Asya/database/nepal.db")
```

```python
query = """
SELECT distinct(s.building_id) AS b_id,
s.*,d.damage_grade
FROM building_structure AS s
INNER JOIN damage_grade AS d ON s.building_id = d.building_id
WHERE s.district_id = 36
"""
print(query)
```

```python
#Read SQL query into a DataFrame using pandas.
df = pd.read_sql(query,conn,index_col="b_id")

df.head()
```
